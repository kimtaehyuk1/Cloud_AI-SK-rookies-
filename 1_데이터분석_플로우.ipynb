# 데이터 수집(획득)

- 4차 산업에 코어(원유)에 해당되는 부분
  - 무료 서비스 -> 데이터 수집 목적이 존재
  - 예외적인 상황
    - 초기에 데이터가 없어도 가능한 방식
    - 강화학습(머신러니으이 학습법중 1개)
      - 플레이를 통해서 데이터가 생성,학습 되는 방식 => 경험을 통해서 학습 -> 초기 데이터없어도 게임의 룰이 중요하다.

## 데이터 종류

- 데이터 종류
  - 정형 데이터
    - **구조화 된 데이터**
    - 대부분 출처가 **관계형데이터베이스**를 통해서 추출된 데이터 확률이 높다.
    - 데이터와 구조정보(스키마)가 분리되어 있다.
  - 비정형 데이터
    - 구조가 없는 데이터
    - 바이너리 데이터
      - 비디오, 오디오, 문서파일
    - NoSQL
      - 빅데이터에서 다루는 형태
  - 반정형 데이터
    - RDBMS(관계형)에서 추출(구성)하진 않았지만, 어느 정도 구조는 가지고 있다.
    - JSON(키-값 형), XML -> 텍스트 기반, 이기종(어디서 쓰나 다 똑같이 이해), 플랫폼독립적 데이터 포멧
    - HTML
    - 반정형은 : 데이터안에 데이터+구조정보(스키마)가 같이 존재

- 정형과 반정형은 데이터안에 데이터+구조정보가 분리 되는지 같이 있는지고, 비정형은 아예 그 구조가 없는거

## 데이터 수집 방식

- 데이터 수집 전략(난이도)
  - 모든 데이터는 그 사용에 관해서 법률적인 검토가 필요하다.(모든 레벨에서)
  - level 1
    - 정형/반정형된 데이터가 대상(일반론)
    - 종류
      - 사내 데이터
      - 공공 데이터(국가기관, 민관, )
      - 경쟁 대회 데이터(캐글, 데이콘, 아레나, 지자체..)
        - 기간이 지나면 미제공 할 수 있음
      - 국책
        - AI허브
      - 바우처
        - 국가에서 데이터 구매시 사용하려고 제공하는 구매비용제공 쿠폰.
          - 데이터 거래소





- 인터넷을 통해서 직접적 수집(레벨1에서는 내가 원하는 데이터를 얻을 수 없었다는 가정!!) 
  - level 2
    - OPEN API를 활용
    - 특정기업의 공개용 API를 이용하여 수집
    - 일일 쿼터량 제한
    - 데이터는 반정형 데이터(출처는 정형 데이터 -> 데이터는 DB에 있는데 내 손에 들어왔을땐 반정형 - json이나 xml)

  - level 3
    - Web Scraping(웹스크래핑)
    - 웹사이트 접속후 어떤 액션도 없이 바로 로딩된 html를 긁어서, 파싱후 데이터 추출.
    - 환율, 위키피디아(자연어구할때), .....
    - BS4(대표 라이브러리)

  - level 4
    - Crawling(크롤링)
    - 사용자의 노력이 들어가서 특정 정보를 접근할 수 있는 사이트가 대상.
      - 로그인, 스크롤, 클릭, ajax...
      - 다른 관점에서는 매크로라고 표현한다.(자동예매,자동댓글,자동글쓰기)
      - 과거 php, js로 많이했고, 현제는 python으로 많이한다.
      - 대표라이브러리 Selenium 

## 자동화

- 데이터 수집 빈도/주기 등등에 따라 매번 사람이 수집할수는 없다
- 자동화를 통해서 알아서 수집 -> 정제-> 적재등등 하나의 플로우로 작동되게 구성
- 방법
  - 1. 파이썬으로 스크립트(목적성)를 작성
  - 2. 하나의 배치파일로 작동되게 환경 구성
  - 3. 자동화 방식 ( os에 종속적 구성)
    - window server
      - **작업 스케줄러**
    - linux
      - **cron**
    - 해당 서버는 통상 클라우드에 구성하여 처리(로컬 PC도 가능함)
- 주의
  - 수집 주기가 짧으면 => 디도스 공격으로 인식할 수 있다.(타켓사이트)
  - 주기를 적절하게 조정 혹은 비번해야하면 프록시(IP 추적을 막기 위해)를 구성해서 우외하여 수집.
    - 특정 IP에서 지속적으로 들어온다는 데이터를 숨기는 용도
  - User Agent라는 요청시 헤더값을 조정하여서(매번 변조하여서) 수집

# 모델 구축

- 통계 모델
- 머신러닝(ML) 모델
  - 지도 학습 ( 정답이 있다.)
    - 분류 ( 이사진은 고양이일 확률이 95%)
    - 회귀 ( 이 집은 xxx원으로 예측된다...->가격예측, 수치값, 범위가 존재)
  - 비지도 학습 ( 정답이 없다)
    - 군집(클러스터링), 넷플릭스 고객 분류
    - 차원축소(데이터압축), 전처리에 적용 될 확률이 높음
  - 준지도 학습
    - 지도 + 비지도 (어떤건 답이 있고 어떤건 답이 없어)
  - 강화 학습
    - 어떤 행동에 따른 리워드(보상)이 주어지면 그 행동을 강화하려는 경향이 보인다. -> 학습에 적용 -> 학습시키고자 하는 대상(에이전트)이 최대 보상을 얻도록 행동을 강화하는 학습법이다. -> 유한게임 환경에서 구현이 가능(무한게임은 쉽지 않다 -> 바둑은 유한게임이고 이런식으로 알파고 학습)
  - 머신러닝은 **알고리즘**을 훈련시켜서 사람이 할 수 없는 일을 해결하고자 함이 목표이다.
  - **알고리즘, 하이퍼파라미터값**
- 딥러닝(DL) 모델
  - 딥러닝 인간의 뉴런구조(신경구조)를 응용하여서 사람이 해결하지 못하는 트정 주제를 해결하기 위해 **인공신경망**을 구성하고 이를 학습을 통해서 해결하는 것을 목표로 한다.
  - **신경망 구조, 가중치(W), 편향(B)을 저장**하고,-> 이렇게 미세조정했을때 이런일 하드라 레시피랑 같음. 필요할 때 가져다 사용하면 언제든지 사용가능
  - 영상, 이미지, 자연어, 생성, ...(주제들)
  - CNN으로 부터 파생, RNN부터 파생, GAN부터 파생.......

# 시스템통합/신규서비스 구현

- 실제 구성한(계획한) 시스템에, 앞에서 구성한 모델을 적용하여 실제 서비스를 구축/운용/유지보수 하는 단계
- MLOps
  - 현재 적용되면서, 정립되고 있는 단계
  - 데이터 분석가, 모델러, DevOps, 등등 담당자가 MLOps로 업무중이다(현재는)
  - 데이터 분석가, 모델러 -> MLOps 로 가는다
  - 특정 포지션으로 취업 후 내부에서 해당 업무로 업그레이드
  - 구성 내용 중심으로 내용
    - 서비스(웹, flask, Django)
    - 데이터
      - 학습용 데이터
      - 검증용 데이터
      - 테스트용 데이터
      - 서비스중 발생하는 고객에 만드는 데이터(활동, 업로드, 서비스 형태 마다 다름)
    - 모델
        - 학습
          - 버전 관리
            - 가중치, 편향, 정확도 같은 지표 관리(대시보드등 관리)
            - 모델 서빙할때 업그레이드, 다운로드가 쉬워야 한다.
    - 환경
        - 서비스가 구동되는 환경 Docker 기반 구축
        - docker에 모델이 세팅된 형태가 아닌 요청시 제공해 주는 형태로 제공해 주는 형태로 구성
        - docker에서 쿠버네티스를 이용한 확장성 및 유연성을 보장(트래픽이 많으면 용량을 키우고...)
    - 기타 환경
      - aws 서비스 환경
      - s3에서는 데이터 업로드, 관리등등 구축
    
  - 고객에게 최적의 서비스를 제공, 관리자는 편하게 관리, 모델을 유연하게 서빙할 수 있는 환경 구축하는 포지션
    - 초기는 심플하게 구성
    - 사용자 늘어나고, 용량이 부족해지면 확장 한다.
